```python
import pandas as pd
import numpy as np

from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_squared_log_error
```

```python
path = 'C:/Users/user/Desktop/PLoSONE/data/'

data = pd.read_csv(path + 'KU_dormitory_PLoS ONE.csv')
data.head()
```

```python
#data_drop = data.drop(columns = 'Date', axis = 1)
#data_drop.head()
```

```python
x_train = data.iloc[:20472]
x_test = data.iloc[20472:]
y_train = data.iloc[:20472]
```

```python
# target 정의 
y = y_train['Consumption']
```

```python
# 데이터 셋 분리
X_train, X_valid, Y_train, Y_valid = train_test_split(x_train, y, test_size=0.3, random_state=42)
```

```python
# GBM
gbm = GradientBoostingRegressor()

params = {'n_estimators': [100, 250, 500],
            'learning_rate': [0.01, 0.05, 0.1],
            'max_depth': [5, 10],
            'loss': ['quantile', 'huber']
}

# GridSEarch
gr = GridSearchCV(gbm, param_grid = params, cv = 5)
gr.fit(X_train, Y_train)

print(gr.best_params_)
```

```python
gbm = GradientBoostingRegressor(random_state=42) # 다시 넣기
gbm.fit(X_train, Y_train)

y_pred = gbm.predict(X_valid)
y_pred
```

```python
# 실제값 
y_predict = gbm.predict(x_test)
y_predict.shape
```
